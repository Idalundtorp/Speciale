```python
FYI_data = OIB_data('FYI', new=False)
FYI_data.TbV_AMSR_all
TB7 = np.array([TbV_AMSR_all[0] for TbV_AMSR_all in FYI_data.TbV_AMSR_all])
TB19 = np.array([TbV_AMSR_all[2] for TbV_AMSR_all in FYI_data.TbV_AMSR_all])
TB37 = np.array([TbV_AMSR_all[4] for TbV_AMSR_all in FYI_data.TbV_AMSR_all])
```

Rotosky

```python
SD_ros = 19.74-555.69*(TB19-TB7)/(TB19+TB7)
```

```python
SD_Kil = 177.01+1.75*TB7-2.80*TB19+0.41*TB37
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from scipy.stats import pearsonr
import math


def RMSE(y_actual, y_predicted):
    MSE = mean_squared_error(y_actual, y_predicted)
    RMSE = math.sqrt(MSE) 
    # print("Root Mean Square Error:", RMSE)
    return RMSE

print(pearsonr(np.sum(FYI_data.sd, axis=1), SD_ros/100))
print(mean_absolute_error(np.sum(FYI_data.sd, axis=1), SD_ros/100))
print(np.sum(np.sum(FYI_data.sd, axis=1)-SD_ros/100)/len(SD_ros/100))
print(r2_score(np.sum(FYI_data.sd, axis=1),SD_ros/100))
print(RMSE(np.sum(FYI_data.sd, axis=1),SD_ros/100))

print(pearsonr(np.sum(FYI_data.sd, axis=1), SD_Cav/100))
print(mean_absolute_error(np.sum(FYI_data.sd, axis=1), SD_Cav/100))
print(np.sum(np.sum(FYI_data.sd, axis=1)-SD_Cav/100)/len(SD_Cav/100))
print(r2_score(np.sum(FYI_data.sd, axis=1),SD_Cav/100))
print(RMSE(np.sum(FYI_data.sd, axis=1),SD_Cav/100))

print(pearsonr(np.sum(FYI_data.sd, axis=1), SD_Kil/100))
print(mean_absolute_error(np.sum(FYI_data.sd, axis=1), SD_Kil/100))
print(np.sum(np.sum(FYI_data.sd, axis=1)-SD_Kil/100)/len(SD_Kil/100))
print(r2_score(np.sum(FYI_data.sd, axis=1),SD_Kil/100))
print(RMSE(np.sum(FYI_data.sd, axis=1),SD_Kil/100))
```
